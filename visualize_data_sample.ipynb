```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
def show_imgs(fullpath):
    images = []
    for img_path in glob.glob(fullpath):
        images.append(mpimg.imread(img_path))

    images = images[:10]

    plt.figure(figsize=(20,10))
    columns = 5
    for i, image in enumerate(images):
        plt.subplot(len(images) / columns + 1, columns, i + 1)
        plt.imshow(image)
    return
```

```python
show_imgs('input/train/images/circle/*.png')
```

```python
show_imgs('input/train/masks/circle/*.png')
```

```python
show_imgs('input/train/images/rectangle/*.png')
```
